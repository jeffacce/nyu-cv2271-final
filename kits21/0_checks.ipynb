```python
for i in tqdm(range(300)):
    case = ROOT / ('case_%05d' % i) / 'imaging.nii.gz'
    img = nib.load(case.as_posix())
    seg = nib.load((ROOT / ('case_%05d' % i) / 'aggregated_MAJ_seg.nii.gz').as_posix())
    x_spacing, y_spacing, z_spacing = -img.affine[0,2], -img.affine[1,1], -img.affine[2,0]
    assert x_spacing > 0
    assert y_spacing > 0
    assert z_spacing > 0
    assert np.isclose(img.affine, seg.affine).all()
    arr = img.get_data()
    print(arr.min(), (arr == arr.min()).sum() / (arr.shape[0] * arr.shape[1] * arr.shape[2]) * 100, arr.max(), (arr == arr.max()).sum() / (arr.shape[0] * arr.shape[1] * arr.shape[2]) * 100)
```
